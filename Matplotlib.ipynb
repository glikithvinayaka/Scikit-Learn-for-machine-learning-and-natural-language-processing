```python
#Import numpy for generating random numbers
import numpy as np
#Import matplot library
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
```

```python
#generate random numbers (10)
random_numbers=np.random.rand(10)
```

```python
#view them
random_numbers
```

```python
#select the style of the plot
style.use('ggplot')
#plot the random number
plt.plot(random_numbers,'b',label='line one', linewidth=2)
#X axis is number of random numbers (index)
plt.xlabel('Range')
#Y axis is actual random number
plt.ylabel('Numbers')
#Title of the plot
plt.title('First Plot')
plt.show()
plt.legend()
```

```python
#Plot with (X,Y)
#Number of users/visitors o the website
web_customers = [123,456,234,566,234,432,543,211,789]
#Time Distribution (hourly)
time_hrs= [7,8,9,10,11,12,13,14,15]
#Select the styl of the plot
style.use('ggplot')
#Plot the websit traffic data (x-axis as hours and y-axis as number of users)
plt.plot(time_hrs,web_customers,'g',linestyle='--',linewidth=2.5, alpha=0.5)
#Set the title of the plot
plt.title('Website Traffic')
#Annotate
plt.annotate('Max',ha='center',va='bottom',xytext=(8,1500),xy=(11,1630),arrowprops={'facecolor':'green'})
#Set label for X-Axis
plt.xlabel('Hrs')
#Set label for Y-Axis
plt.ylabel('Number of users')
plt.show()
```

```python
#Plot with (X,Y)
#Number of users/visitors to the website on monday
web_monday = [123,456,234,566,234,432,543,211,789]
#Number of users/visitors to the website on tuesday
web_tuesday = [321,654,432,665,432,234,345,112,987]
#Number of users/visitors to the website on wednesday
web_wednesday = [234,564,342,656,342,324,435,121,897]
#Time Distribution (hourly)
time_hrs= [7,8,9,10,11,12,13,14,15]
#Select the styl of the plot
style.use('ggplot')
#Plot the websit traffic data (x-axis as hours and y-axis as number of users for different days)
plt.plot(time_hrs,web_monday,'r',label='monday',linewidth=1)
plt.plot(time_hrs,web_tuesday,'g',label='tuesday',linewidth=1.5)
plt.plot(time_hrs,web_wednesday,'b',label='wednesday',linewidth=2)
#Set title
plt.title('Web Traffic on Different Days')
plt.xlabel('Hrs')
plt.ylabel('Number of users')
plt.show()
```
